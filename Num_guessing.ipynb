```python
import random

num_guess=random.randint(1,20)
num1=int(input('Enter your 1st choice number (1 to 20): '))
if num1==num_guess:
    print('You won')

elif num1<num_guess:
  print('Low guess')

else:
  print('High guess')

if num1!=num_guess:
  num2=int(input('Enter your 2nd choice number (1 to 20):'))

  if num2==num_guess:
    print('You won')

  elif num2<num_guess:
    print('Low guess')

  else:
    print('High guess')

if num1!=num_guess and num2!=num_guess:
  print(f'Both guess failed! The right number was {num_guess}')

```
